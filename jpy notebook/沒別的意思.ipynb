```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1=np.arange(0,1.01,0.01)
y1=-2*(x1-1)*x1

x2=np.arange(1,2.01,0.01)
y2=-2*(x2-2)*(x2-1)

x3=np.arange(2,3.01,0.01)
y3=-15*(x3-2)*(x3-3)

x4=np.arange(3,4.01,0.01)
y4=-2*(x4-3)*(x4-4)

x5=np.arange(4,5.01,0.01)
y5=-2*(x5-4)*(x5-5)

x=np.arange(-np.pi,0,0.01)
x6=np.cos(x)*2.5+2.5
y6=np.sin(x)*4

plt.plot(x1,y1)
plt.plot(x2,y2)
plt.plot(x3,y3)
plt.plot(x4,y4)
plt.plot(x5,y5)
plt.plot(x6,y6)
```
